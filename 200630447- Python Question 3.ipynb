```python
#These are the libraries we are going to use
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_columns', None)
import warnings
warnings.filterwarnings('ignore')
```

```python
#To import data set
data1 = pd.read_csv('2006.csv')
data2 = pd.read_csv('2007.csv')
```

```python
#merging method should have equal no. of columns
data = pd.concat([data1,data2])
```

```python
data.head()
```

```python
airportdata = pd.read_csv('airports.csv')
```

```python
airportdata.head()
```

Data merging

```python
#Merging with 2 diff columns and only using some columns in airport data
newdataset = pd.merge(data,airportdata[['iata','city']],left_on = 'Origin', right_on = 'iata',how = 'left')
```

```python
newdataset.head()
```

```python
#How to rename an existing column
newdataset = newdataset.rename(columns = {'city':'Origin_city'})
```

```python
newdataset.head()
```

```python
#Merging with 2 diff columns and only using some columns in airport data
newdataset = pd.merge(newdataset,airportdata[['iata','city']],left_on = 'Dest', right_on = 'iata',how = 'left')
```

```python
#How to rename an existing column
newdataset = newdataset.rename(columns = {'city':'Dest_city'})
```

```python
newdataset.head()
```

```python
newdataset['location'] = newdataset['Origin_city'] + '-'+ newdataset['Dest_city']
```

These are the most number of flights compared to Origin and Destination

```python
newdataset.location.value_counts().head(10)
```

```python
locationwiselist = ['Boston-New York','New York-Boston','Chicago-New York','New York-Chicago','Kahului-Honolulu','Honolulu-Kahului','New York-Arlington','Arlington-New York','Houston-Dallas','Dallas-Houston']
```

```python
#isin => checks list and checks values in the list(matching values)
locationwisedata = newdataset[newdataset['location'].isin(locationwiselist)]
```

```python
locationwisedata.head()
```

```python
#index counts the number of variables
locationwisedata['ind'] = locationwisedata.index
```

```python
locationwisedata1 = locationwisedata.groupby(['Year','location']).agg(numberofflights = ('ind','count')).reset_index()
```

```python
data.Origin.value_counts().head(10)
```

# Plottings for Question 3

```python
plot = sns.catplot(data = locationwisedata1, kind = 'bar', x = 'location', y = 'numberofflights', hue = 'Year', aspect = 2)
#rotates the bottom labels
plot.set_xticklabels(rotation = 90)
```

```python
locationwisedata['MonthAndYear'] = locationwisedata['Year'].astype(str) +'-'+ locationwisedata['Month'].astype(str)
```

```python
locationwisedata2 = locationwisedata.groupby(['MonthAndYear','location']).agg(numberofflights = ('ind','count')).reset_index()
```

```python
plot = sns.catplot(data = locationwisedata2, kind = 'bar', x = 'MonthAndYear', y = 'numberofflights', hue = 'location', aspect = 2)
#rotates the bottom labels
plot.set_xticklabels(rotation = 90)
```
