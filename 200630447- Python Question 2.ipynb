```python
#These are the libraries we are going to use
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_columns', None)
import warnings
warnings.filterwarnings('ignore')
```

```python
#To import data set
data1 = pd.read_csv('2006.csv')
data2 = pd.read_csv('2007.csv')
planedata = pd.read_csv('plane-data.csv')
```

```python
#merging method should have equal no. of columns
data = pd.concat([data1,data2])
```

```python
data.head(10)
```

```python
planedata.head(10)
```

```python
#Renaming tail num column
planedata = planedata.rename(columns = {'tailnum':'TailNum'})
```

```python
#To merge datasets which have the same columns
data = pd.merge(data,planedata,on = 'TailNum', how = 'left')
```

```python
#[] used as subset might be a list
tailnumbermatched = data.dropna(subset =['year'])
```

```python
tailnumbermatched.head()
```

```python
#Making a new data set with selected columns
checkingdelays = tailnumbermatched.groupby(['year']).agg(ArrivalDelay = ('ArrDelay','sum'),DepartureDelay = ('DepDelay','sum')).reset_index()
```

```python
checkingdelays['alldelay'] = checkingdelays['ArrivalDelay']+checkingdelays['DepartureDelay']
```
