Creating date/time variable

```python
#Taking minutes separately
Minutes = data.CRSDepTime.astype(str).str[-2:]
Minutes
```

```python
# Taking hours separately
Hours = data.CRSDepTime.astype(str).str[:-2]
Hours
```

```python
Hours[Hours == ""] = "0"
```

```python
DayofTheMonth = data.DayofMonth.astype(str)

#apply used to remove decimal numbers from date format(only integers)
DayofTheMonth = DayofTheMonth.apply(lambda x: '{0:0>2}'.format(x))
```

```python
DayofTheMonth
```

```python
data["DateTimeCRSDepTime"] = pd.to_datetime(
            DayofTheMonth.astype(str) + "/" + data.Month.astype(str) + "/" + data.Year.astype(str) + "," + Hours + ":" + Minutes, 
            format='%d/%m/%Y,%H:%M')
data["DateTimeCRSDepTime"]
```

```python
data.TailNum.value_counts()
```

```python
data["alldelay"] = data.ArrDelay + data.DepDelay
```

```python
#shift is used to take the previous row
data["PreviousDelay"] = data.groupby(['TailNum', 'DateTimeCRSDepTime']).shift(-1).alldelay
```

Plottings for question 4

```python
plt.scatter(data["PreviousDelay"], data["alldelay"])
plt.xlabel('PreviousDelay')
plt.ylabel('CurrentDelay')
plt.show()
```

```python
data['HasCurrentDelay'] = np.where(data['alldelay'] > 0, 1, 0)
data['HasPreviousDelay'] = np.where(data['PreviousDelay'] > 0, 1, 0)
```

```python
#crosstab makes a count based on 2 features
Crosstab = pd.crosstab(data.HasPreviousDelay, data.HasCurrentDelay, rownames=["Previous Delay"], colnames=["Currrent Delay"])

Crosstab 
```
